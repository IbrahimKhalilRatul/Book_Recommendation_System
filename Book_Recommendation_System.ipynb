```python
import pandas as pd

# Replace 'path/to/your/dataset.csv' with the actual path to your dataset
file_path = '/content/drive/MyDrive/archive (5)/books.csv'

data = pd.read_csv(file_path, on_bad_lines='skip')  # Updated to handle bad lines

print(data.head())
```

```python
print(data.columns)
```

```python
data["combined_text"] = data["title"] + " " + data["authors"]
```

```python
!pip install sentence-transformers
from sentence_transformers import SentenceTransformer # Imports the necessary class

# Load a pre-trained sentence transformer model
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
# Generate embeddings for book descriptions
book_embeddings = model.encode(data["combined_text"].tolist())
```

```python
from sklearn.metrics.pairwise import cosine_similarity # Import cosine_similarity

# Compute cosine similarity between books
similarity_matrix = cosine_similarity(book_embeddings)
```

```python
def get_recommendations(book_title, top_n=5):
    book_index = data[data["title"] == book_title].index[0]
    similarity_scores = list(enumerate(similarity_matrix[book_index]))
    similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)

    # Get top_n most similar books, excluding the book itself and duplicates
    top_books = []
    for i, score in similarity_scores[1:]:  # Start from index 1 to exclude the book itself
        recommended_book_title = data['title'][i]
        if recommended_book_title not in top_books and recommended_book_title != book_title:  # Corrected condition
            top_books.append(recommended_book_title)
            if len(top_books) == top_n:
                break  # Stop when we have enough recommendations

    return top_books
```

```python
# Test the recommendation system
book_title = "The Great Gatsby"  # Replace with a book title from your dataset
recommendations = get_recommendations(book_title)
print(f"Recommendations for '{book_title}':")

# Instead of directly indexing with the list, filter the DataFrame:
recommended_books = data[data['title'].isin(recommendations)]

# Display the title and author on separate lines:
for index, row in recommended_books.iterrows():
    print(f"Title: {row['title']}")
    print(f"Author: {row['authors']}")
    print("---")
```
